```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
var = 8
#x-axis ranges from -3 and 3 with .001 steps
x = np.arange(-90, 90, 0.01)
# generate 10 random numbers and softmax them
coef = np.random.randn(20)
coef = np.exp(coef)
coef = coef / np.sum(coef)

# generate 20 randome number in range (-90,90)
rand_seed = np.random.uniform(-90, 90, 20)
# generate 10 pdf of norm with random mean in range (-90,90) and standard deviation 1
y = []
for i in range(20):
    temp = norm.pdf(x, rand_seed[i], 1)*coef[i]
    y.append(temp)
    # plt.plot(x, y[i])
#plot y and the summation of y
plt.plot(x, np.sum(y, axis=0))
```

```python
# plot the corresponding cdf base on sum of y
y_cdf = []
for i in range(20):
    temp = norm.cdf(x, rand_seed[i], 1)*coef[i]
    y_cdf.append(temp)
plt.plot(x, np.sum(y_cdf, axis=0))
    
```
